## Creating Functions to -

1.Get the topic page

2.Get the topic information (user name, repo name, star count and repo url)

3.Get the topic tags and star tags

```python
#Global variable
base_url = 'https://github.com/'
```

```python
#Function for converting stars from string to integer
def parse_star_count(stars):
    stars=stars.strip()
    if stars[-1]=='k':
        return int(float(stars[:-1])*1000)
    return(int(stars))
```

```python
#Function to get the topic page info
def get_topic_page(topic_url):
    
    ''' This function takes topic ural as an input and returns the parsed html page as a doc variable'''
    
    response = requests.get(topic_url)
    if response.status_code != 200:
        raise Exception('Failed to load page {}'.format(topic_url))
    topic_doc = BeautifulSoup(response.text, 'html.parser')
    return topic_doc

#Function to get the repo info 
def get_repo_info(repo_tag,star_tag):
    
    '''This function takes "repo_tag" which is used to get repos information and "star_tag" which is used to get the 
    count of stars on the repo and returns "username","repo_name", "star count", "repo url"'''
    
    a_tag = repo_tag.find_all('a')
    username = a_tag[0].text.strip()
    repo_name = a_tag[1].text.strip()
    repo_url =  base_url + a_tag[1]['href']
    stars = parse_star_count(star_tag.text.strip())
    return username, repo_name, stars, repo_url

#Function for getting repo info info
def get_topic_repos(topic_doc):
    
    '''This function takes topic_doc as an input and creates a dataframe containing information about username,repo name
    star count repo url'''
    
    repo_tags = topic_doc.find_all('h3',class_= 'f3 color-fg-muted text-normal lh-condensed')
    star_tag = topic_doc.find_all('span',class_='Counter js-social-count')
    topic_repos_dict = { 'username': [], 'repo_name': [], 'stars': [],'repo_url': []}

    for i in range(len(repo_tags)):
        repo_info = get_repo_info(repo_tags[i], star_tag[i])
        topic_repos_dict['username'].append(repo_info[0])
        topic_repos_dict['repo_name'].append(repo_info[1])
        topic_repos_dict['stars'].append(repo_info[2])
        topic_repos_dict['repo_url'].append(repo_info[3])
        
    return pd.DataFrame(topic_repos_dict)

#Function to scrape topic
def scrape_topic(topic_url, path):
    
    ''' This fucntion takes topic url and path as input and creates csv file for each topic info'''
    
    if os.path.exists(path):
        print("The file {} already exists. Skipping...".format(path))
        return
    topic_df = get_topic_repos(get_topic_page(topic_url))
    topic_df.to_csv(path, index=None)
```

## Creating functions to -

1.Get the list of topics from the topic page

2.Get the list of top repos from the topics

3.Create a csv file for each topic with top repos information

```python
#Function to get topic titles
def get_topic_titles(doc):
    
    '''This function returns topic titles'''
    
    selection_class = 'f3 lh-condensed mb-0 mt-1 Link--primary'
    topic_title_tags = doc.find_all('p', {'class': selection_class})
    topic_titles = []
    for tag in topic_title_tags:
        topic_titles.append(tag.text)
    return topic_titles

#Function to get topic description
def get_topic_descs(doc):
    
    '''This function returns topic descriptions'''
    
    desc_selector = 'f5 color-text-secondary mb-0 mt-1'
    topic_desc_tags = doc.find_all('p',class_ ='f5 color-fg-muted mb-0 mt-1')
    topic_descs = []
    for tag in topic_desc_tags:
        topic_descs.append(tag.text.strip())
    return topic_descs

#Function to get topic url
def get_topic_urls(doc):
    
     '''This function returns topic urls'''
    
    topic_link_tags = doc.find_all('a', {'class':'no-underline flex-grow-0'})
    topic_urls = []
    base_url = 'https://github.com'
    for tag in topic_link_tags:
        topic_urls.append(base_url + tag['href'])
    return topic_urls
    
#Fucntion to scrape topic information
def scrape_topics():
    
    '''This function gives topic information with title,description,url'''
    
    topics_url = 'https://github.com/topics'
    response = requests.get(topics_url)
    if response.status_code != 200:
        raise Exception('Failed to load page {}'.format(topic_url))
    topics_dict = {
        'title': get_topic_titles(doc),
        'description': get_topic_descs(doc),
        'url': get_topic_urls(doc)
    }
    return pd.DataFrame(topics_dict)
```

```python
def scrape_topics_repos():
    
    '''This function returns the csv file containing top repo info for each individual topic'''
    
    print('Scraping list of topics')
    topics_df = scrape_topics()
    
    os.makedirs('data', exist_ok=True)
    for index, row in topics_df.iterrows():
        print('Scraping top repositories for "{}"'.format(row['title']))
        scrape_topic(row['url'], 'data/{}.csv'.format(row['title']))
```

```python
#Calling the function to scrape all the topic information and save it as a csv file for each individual top
scrape_topics_repos()
```
